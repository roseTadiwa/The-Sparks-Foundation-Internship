```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd


df = pd.read_csv('C:/Users/ROSEMARY/OneDrive/Desktop/New datasets/Global Terrorism - START data/terrorism.csv',
                 encoding='latin-1',
                 low_memory=False)

# Determine the columns with mixed data types
mixed_type_cols = [col for col, dtype in df.dtypes.items() if dtype == 'object']

# Create a dictionary of data types to use
dtypes = {col: 'object' for col in mixed_type_cols}

# Read the file again with the specified data types
df = pd.read_csv('C:/Users/ROSEMARY/OneDrive/Desktop/New datasets/Global Terrorism - START data/terrorism.csv',
                 encoding='latin-1',
                 dtype=dtypes)
```

```python
print(df.head())
```

```python
# Examining the distribution of attacks over time
df.groupby('Year')['iday'].count().plot(kind='bar')
plt.title('Number of Terrorist Attacks by Year')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
# Identifying the countries with the most attacks
top_countries = df['country_txt'].value_counts().head(10)
top_countries.plot(kind='bar')
plt.title('Top 10 Countries by Number of Terrorist Attacks')
plt.xlabel('Country')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
# Analyzing the types of attack methods used
attack_types = df['attacktype1_txt'].value_counts()
attack_types.plot(kind='bar')
plt.title('Types of Terrorist Attack Methods')
plt.xlabel('Attack Type')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
# Investigating the targets of terrorist attacks
target_types = df['targtype1_txt'].value_counts()
target_types.plot(kind='bar')
plt.title('Types of Terrorist Attack Targets')
plt.xlabel('Target Type')
plt.ylabel('Number of Attacks')
plt.show()
```

```python

# Identifying the groups responsible for the attacks
groups = df['gname'].value_counts().head(10)
groups.plot(kind='bar')
plt.title('Top 10 Terrorist Groups')
plt.xlabel('Terrorist Group')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
target_type_counts = df['targtype1_txt'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(target_type_counts, labels=target_type_counts.index, autopct='%1.1f%%')
plt.title('Proportion of Terrorist Attacks by Target Type')
plt.axis('equal')
plt.show()
```

```python
target_type_counts = df['success'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(target_type_counts, labels=target_type_counts.index, autopct='%1.1f%%')
plt.title('Proportion of Terrorist Attacks by Target Type')
plt.axis('equal')
plt.show()
```

```python
target_type_counts = df['ransom'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(target_type_counts, labels=target_type_counts.index, autopct='%1.1f%%')
plt.title('Proportion of Terrorist Attacks by Ransom')
plt.axis('equal')
plt.show()
```

```python
# Scatter plot of the number of attacks vs. the number of casualties
plt.figure(figsize=(8, 6))
plt.scatter(df['nkill'], df['nwound'])
plt.title('Terrorist Attacks: Number of Kills vs. Wounds')
plt.xlabel('Number of Kills')
plt.ylabel('Number of Wounds')
plt.show()
```
